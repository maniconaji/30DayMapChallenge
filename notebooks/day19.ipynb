```python
import ee
import os
import geemap
import ipywidgets as widgets
```

```python
# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize()
```

```python
def maskS2clouds(image):
    qa = image.select('QA60')
    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloudBitMask = 1 << 10;
    cirrusBitMask = 1 << 11;
    # Both flags should be set to zero, indicating clear conditions.
    mask1 = qa.bitwiseAnd(cloudBitMask).eq(0)
    mask2 = qa.bitwiseAnd(cirrusBitMask).eq(0)
    return image.updateMask(mask1).updateMask(mask2).divide(10000)
```

```python
xmin, xmax, ymin, ymax =-79.121686, -78.593696, -33.792880, -33.513537
xcenter, ycenter = -78.862903, -33.631043
roi = ee.Geometry.Rectangle(coords = [[xmin, ymin], [xmax, ymax]])
```

```python
Map = geemap.Map()
```

```python
dataset = ee.ImageCollection('COPERNICUS/S2')\
            .filterDate('2018-01-01', '2018-06-30')\
            .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))\
            .map(maskS2clouds)
```

```python
print(dataset.size().getInfo())
```

```python
rgbVis = dict(min = 0.0, max = 0.3, bands = ['B4', 'B3', 'B2'])
Map.setCenter(xcenter, ycenter, 11)
Map.addLayer(dataset.median(), rgbVis, 'RGB')
Map
```
